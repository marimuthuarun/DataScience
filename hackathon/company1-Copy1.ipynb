```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
company=pd.read_csv('company.csv')
company.tail(100)
```

```python
#display TCS employees
company.fillna( method ='ffill', limit = 1, inplace = True) 
company[company['Company'].str.match('TCS')]
```

```python
#Display males working in TCS
company[company['Company'].str.match('TCS') & (company['Gender']==1)]
```

```python
#display working in TCS and Chennai place
company[company['Company'].str.match('TCS') & (company['Place']=='Chennai')]
```

```python
#who are all working more than 5000 salary
company[company['Salary']>5000]
```

```python
#more than 5000 salary and age above 45
company[(company['Salary']>5000) & (company['Age']>45)]
```

```python
company.describe()
```

```python
company.count(axis='columns')
```

```python
#mean of the age
company.loc[:,'Age'].mean()
```

```python
company.plot(x='Company', y='Age',kind='line')
plt.show()
```

```python
plt.figure(figsize = (16,6))
#add a title
plt.title("Age VS Salary")
sns.lineplot(data = company[['Age','Salary']])
plt.xlabel('Age')
```

```python
plt.figure(figsize = (10,6))
plt.title("Plot for Age VS Salary")

#defining bar plot here
sns.barplot(x = company['Age'],  y = company['Salary'])
```
